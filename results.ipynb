```python
import pickle
import matplotlib.pyplot as plt
```

```python
def get_data(data_file):
    data = open(data_file, "rb")
    data = pickle.load(data)
    return data
```

```python
lstm_train_acc = get_data("results/lstm_train_acc.pkl")
lstm_train_loss = get_data("results/lstm_train_loss.pkl")
lstm_val_acc = get_data("results/lstm_val_acc.pkl")
lstm_val_loss = get_data("results/lstm_val_loss.pkl")
transformer_train_acc = get_data("results/transformer_train_acc.pkl")
transformer_train_loss = get_data("results/transformer_train_loss.pkl")
transformer_val_acc = get_data("results/transformer_val_acc.pkl")
transformer_val_loss = get_data("results/transformer_val_loss.pkl")
```

```python
def plot_data(model1: dict, model2: dict, title: str, ylabel: str) -> None:
    plt.figure(figsize=(8, 6))
    plt.plot(list(model1.keys()), list(model1.values()), label="LSTM")
    plt.plot(list(model2.keys()), list(model2.values()), label="Transformer")
    plt.xlabel("Epochs")
    plt.ylabel(ylabel)
    plt.title(label=title, fontweight="bold")
    plt.legend()
    plt.grid()
    plt.savefig("results/img/" + title + ".png")
    plt.show()
```

```python
plot_data(
    model1=lstm_train_acc,
    model2=transformer_train_acc,
    title="Training Accuracy",
    ylabel="Accuracy"
)
```

```python
plot_data(
    model1=lstm_val_acc,
    model2=transformer_val_acc,
    title="Validation Accuracy",
    ylabel="Accuracy"
)
```

```python
plot_data(
    model1=lstm_train_loss,
    model2=transformer_train_loss,
    title="Training Loss",
    ylabel="BCE Loss"
)
```

```python
plot_data(
    model1=lstm_val_loss,
    model2=transformer_val_loss,
    title="Validation Loss",
    ylabel="BCE Loss"
)
```
